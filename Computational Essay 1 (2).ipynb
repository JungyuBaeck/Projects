Background: tables of stellar brightness list the flux density for stars of designated magnitude. In observational astronomy, one of the most fundamental questions we ask about any celestial object is: how much light does it emit? While brightness is often expressed in magnitudes or flux units, a deeper physical understanding comes from thinking in terms of photons, the individual quanta of light. Since every telescope detector counts photons, knowing how many photons arrive from a source at a given wavelength is crucial for designing observations, estimating exposure times, and interpreting data. This essay presents a computational approach to calculating the photon arrival rate



In observational astronomy, stellar brightness is often listed as **flux density per unit frequency** $F_\nu$ in units as 
$$
{W\,m^{-2}\,Hz^{-1}}
$$
However, many physical calculations and analyses require the ***flux density per unit wavelength** $F_\lambda$ in units as
$$
{erg\,s^{-1}\,cm^{-2}\,\text{Å}^{-1}}
$$


We will need to convert $F_\nu$ into $F_\lambda$. We do it this way:

We start from the equality of flux densities:

$$
F_\nu\, d\nu = F_\lambda\, d\lambda
$$

Using the relation between frequency and wavelength:

$$
\nu = \frac{c}{\lambda}
$$

we get the derivative:

$$
\frac{d\nu}{d\lambda} = -\frac{c}{\lambda^2}
$$

Thus, substituting back into the flux equation:

$$
F_\lambda = F_\nu \frac{d\nu}{d\lambda} = F_\nu \frac{c}{\lambda^2}
$$

We have:
$$
F_\nu = F_\lambda \frac{\lambda^2}{c}
$$

Since we want $F_\lambda$ in cgs units we include conversion factors:
$${W} = 10^{7}\,\mathrm{erg\,s^{-1}}$$

$${m^2} = 10^{4}\,\mathrm{cm^2}$$

$${m} = 10^{10}\,\text{Å}$$



The following function does it

```python
# Define function.

def convert_fnu(fnu,w) :
    """
    Returns a value of F-lambda given F-nu
    and a wavelength.
    
    Arguments (required)
    1. F-nu in units of Watt per square m per Hz
    2. Wavelength in Angstrom
    
    Returns
    
    F-lambda in units of erg per second per square 
    cm per Angstrom
    """
    c = 3.0E+8  # m/s
    
    lam_m = w * 1e-6 # Convert wavelength from microns to meters
    fnu_si = fnu * 1e-26 # Convert Fν from Jy to W m^-2 Hz^-1 (1 Jy = 1e-26 W m^-2 Hz^-1)
    
    result = fnu_si * c / (lam_m ** 2)
    result = result * 1.0E+7 / 1.0E+4 / 1.0E+10 
    # Convert to erg s^-1 cm^-2 Å^-1
    # 1 W = 10^7 erg/s, 1 m² = 10^4 cm², 1 m = 10^10 Å
    return result
```

Astronomical detectors count photons, not directly energy. Therefore, knowing the photon arrival rate per unit wavelength is very useful. The energy of a single photon is given by:
$$
E_\gamma = \frac{hc}{\lambda}
$$
The photon rate per unit area per unit wavelength is then:
$$
N_\lambda = \frac{F_\lambda}{E_\gamma} = \frac{F_\lambda \, \lambda}{h c}
$$


$F_\lambda$ = flux density (erg s⁻¹ cm⁻² Å⁻¹)

$\lambda$ = wavelength (Å)

h = Planck’s constant = 6.626×$10^{-27}$ erg·s

c = 2.998x$10^{18}$ Å/s

The following function does it

```python
def photon_rate(Flambda, w):
    """
    Calculate photon rate (photons/s/cm^2/Å)

    Parameters
    ----------
    Flambda : float
        Flux density (erg/s/cm^2/Å)
    w : float
        Wavelength (microns)

    Returns
    -------
    N_lambda : float
        Photon arrival rate per Å
    """
    h = 6.626E-27       # erg·s
    c = 2.998E+18       # Å/s
    lam_A = w * 1.0E+4  # microns -> Å
    return Flambda * lam_A / (h * c)
```

#### Problem 2 -- How Bright is the Night Sky?

##### Instructions:

    How much brighter is the sky at quarter moon (phase = 7) compared to new

We want to compare how much brighter the night sky becomes as the Moon's phase changes. The table gives sky brightness in magnitudes per square arcsecond, which means a 1"x1" patch of sky is as bright as a star of that magnitude. 

We will compute, for each filter (U, B, V, R, I): 

    -The brightness factor at quarter moon (phase 7) to new moon (phase 0)
    -The brightness factor at full moon (phase 14) to new moon (phase 0)
    

Brightness in astronomy uses a logarithmic magnitude scale. If $m_1$ and $m_2$ are magnutudes and $F_1$, $F_2$ are the corresponding fluxes (brightness): 

$\frac{F_1}{F_2} \;=\; 10^{0.4\,(m_2 - m_1)}$


So for a given filter: 


$\text{Brightness ratio} \;=\; \frac{F_{\text{phase}}}{F_{\text{new}}}
\;=\; 10^{0.4\,(m_{\text{new}} - m_{\text{phase}})}$


Smaller magnitude $\Rightarrow$ brighter sky (because it's a logarithmic, inverted scale). 

```python
import numpy as np
import pandas as pd
```

```python
def brightness_ratio(m_new: float, m_phase: float) -> float:
    """
    Return the multiplicative brightness factor of the sky
    at a given phase relative to new moon (phase=0),
    using the astronomical magnitude–flux relation.

    Parameters
    ----------
    m_new : float
        Sky magnitude at new moon (mag/arcsec^2).
    m_phase : float
        Sky magnitude at some other lunar phase (mag/arcsec^2).

    Returns
    -------
    float
        Brightness ratio F_phase / F_new (dimensionless).
    """
    # Magnitude difference formula: F1/F2 = 10^(0.4 * (m2 - m1))
    # Here: F_phase / F_new = 10^(0.4 * (m_new - m_phase))
    return 10 ** (0.4 * (m_new - m_phase))
```

```python
# Filters in order
filters = ["U", "B", "V", "R", "I"]
```

```python
# Magnitudes (mag/arcsec^2) at each phase, by filter (order must match `filters`)
mags_new   = [22.0, 22.7, 21.8, 20.9, 19.9]  # phase 0 (new)
mags_quart = [19.9, 21.6, 21.4, 20.6, 19.7]  # phase 7 (quarter)
mags_full  = [17.0, 19.5, 20.0, 19.9, 19.2]  # phase 14 (full)
```

```python
# Computing brightness ratios with list comprehensions
ratio_quarter_over_new = [brightness_ratio(mn, mq) for mn, mq in zip(mags_new, mags_quart)]
ratio_full_over_new    = [brightness_ratio(mn, mf) for mn, mf in zip(mags_new, mags_full)]
```

```python
# Computing magnitude differences for intuition
delta_m_quarter = [mn - mq for mn, mq in zip(mags_new, mags_quart)]
delta_m_full    = [mn - mf for mn, mf in zip(mags_new, mags_full)]
```

```python
# Assembling a readable table
df = pd.DataFrame({
    "Filter": filters,
    "New Moon (mag)": mags_new,
    "Quarter Moon (mag)": mags_quart,
    "Full Moon (mag)": mags_full,
    "Δm (Quarter - New)": delta_m_quarter,   # positive means quarter is brighter
    "Δm (Full - New)": delta_m_full,         # positive means full is brighter
    "Quarter/New Ratio": ratio_quarter_over_new,
    "Full/New Ratio": ratio_full_over_new
})
```

```python
# Rounds ratios for display; keep delta m to 2 decimals for readability
df_rounded = df.copy()
df_rounded["Δm (Quarter - New)"] = df_rounded["Δm (Quarter - New)"].round(2)
df_rounded["Δm (Full - New)"]    = df_rounded["Δm (Full - New)"].round(2)
df_rounded["Quarter/New Ratio"]  = df_rounded["Quarter/New Ratio"].round(2)
df_rounded["Full/New Ratio"]     = df_rounded["Full/New Ratio"].round(2)

df_rounded
```

Using the assignment's given sky brightness table, the sky becomes roughly 1.2-6.9 times brighter at quarter moon (depending on the filter) and up to 100 times brighter at full moon in the U band, with smaler but still significant brightening in B and much smaller in R and I. In astronomy, magnitudes are derived from flux densities using

$m = -2.5 \log_{10}\!\left(\frac{F_\nu}{F_{\nu,0}}\right),$

where $F_\nu$ is the source flux density and $F_{\nu,0}$ is the zero magnitude flux for that band. 

We're told that the star has the same $F_\nu$ in every filter: 

$F_\nu = 0.0364\ \text{mJy}$. 

We'll use the zero magnitude fluxes (in Jy) from Problem 1's table. 

First we must convert mJy to Jy: 
    
1 mJy = 10⁻³ Jy


So $F_\nu = 0.0364\,\text{mJy} = 3.64\times10^{-5}\,\text{Jy}$.

Next, we'll compute the magnitudes

$m = -2.5\log_{10}\!\left(\frac{3.64\times10^{-5}}{F_{\nu,0}}\right)$

for each band. 

Third, we must compare to the sky. Using the table from problem 2, we can find the magnitudes. We'll use the brightness ratio: 

$\frac{F_\text{star}}{F_\text{sky}} = 10^{0.4\,(m_\text{sky}-m_\text{star})}$.

We'll find this at the new and full moon for each filter. 

```python
#known constants
F_star_mJy = 0.0364                  # given
F_star = F_star_mJy * 1e-3           # convert to Jy

filters = ["U", "B", "V", "R", "I"]
F_nu0 = [1810, 4260, 3640, 3080, 2550]  

# Sky brightness magnitudes
sky_new  = [22.0, 22.7, 21.8, 20.9, 19.9]
sky_full = [17.0, 19.5, 20.0, 19.9, 19.2]
```

```python
# magnitudes of the star
m_star = [-2.5 * np.log10(F_star / F0) for F0 in F_nu0]
```

```python
# brightness ratio vs sky
def flux_ratio(m_obj, m_sky):
    """Return F_obj / F_sky using magnitude difference."""
    return 10 ** (0.4 * (m_sky - m_obj))

ratio_new  = [flux_ratio(ms, mn) for ms, mn in zip(m_star, sky_new)]
ratio_full = [flux_ratio(ms, mf) for ms, mf in zip(m_star, sky_full)]
```

```python
# assembling table
df3 = pd.DataFrame({
    "Filter": filters,
    "Fν0 (Jy)": F_nu0,
    "Star mag": np.round(m_star, 2),
    "Sky (New) mag": sky_new,
    "Sky (Full) mag": sky_full,
    "Star/New Ratio": np.round(ratio_new, 2),
    "Star/Full Ratio": np.round(ratio_full, 2)
})

df3
```

**Observe this object using the 1.3m diameter telescope at MDM.
How fast will photons arrive from this object in the V filter, which has a width of 800 Angstroms?**

Photons hitting detector per second:
$$
N_{\mathrm{det}} = \Phi_{\star,\mathrm{band}} \times A \times \eta
$$
To get  $\Phi_{\star,\mathrm{band}}$, 
We have $F_\nu = F_\lambda \frac{\lambda^2}{c}$, rewrite as $f_\lambda = \frac{f_\nu \lambda^2}{c}$
,and $E_\gamma = \frac{hc}{\lambda}$ from problem 1 to convert V zero point to $F_\lambda$ and energy per photon at $\lambda$.

For the photon flux for a V=0 star:
$$
\Phi_0(\lambda) = \frac{f_{\lambda,0}}{E_\gamma}
$$

For the photon flux for Kepler-90 of magnitude 13.896:
$$
\Phi_{\star}(\lambda) = \Phi_{0} \times 10^{-0.4m_V}
$$


**Now we can calculate photons hitting detector per second with these parameters**:


Kepler-90 V magnitude: $m_V = 13.896$

central wavelength for V band: $\lambda = 5500\,\text{Å} = 5.5 \times 10^{-7}\,\mathrm{m}$

V zero point (spectral flux density):$f_{\nu,0} = 3631~\mathrm{Jy} = 3631 \times 10^{-26}\,\mathrm{W\,m^{-2}\,Hz^{-1}}$

Planck’s constant: $h = 6.626 \times 10^{-34}\,\mathrm{J\,s}$

Speed of light: $c = 2.998 \times 10^{8}\,\mathrm{m\,s^{-1}}$

Filter width: $\Delta\lambda = 800\,\text{Å}$

Telescope diameter: $D = 1.3\,\mathrm{m}$

Collecting area: $A = \pi \left( \frac{D}{2} \right)^2$

System efficiency: $\eta = 0.7$

```python
import numpy as np

m_V = 13.896          # V magnitude of Kepler-90
lambda_c = 5500e-10   # central wavelength in m
delta_lambda = 800    # filter width in Å
eta = 0.7             # system efficiency
D = 1.3               # telescope diameter in m
h = 6.626e-34         # Planck constant in J·s
c = 2.998e8           # speed of light in m/s

# zero-point flux density for V band
fnu_0 = 3631e-26 # in W m^-2 Hz^-1
f_lambda_0 = fnu_0 * c / (lambda_c**2) * 1e-10  # in W m^-2 Å^-1

# Photon energy at central wavelength
E_gamma = h * c / lambda_c

# Photon flux for a V=0 star photons in s^-1 m^-2 Å^-1
phi_0 = f_lambda_0 / E_gamma

# Photon flux for Kepler-90
phi_star_perA = phi_0 * 10**(-0.4 * m_V)

# Integrate over filter width
phi_star = phi_star_perA * delta_lambda  # photons in s^-1 m^-2

# Telescope collecting area
A = np.pi * (D / 2)**2  # in m^2

# Detected photon rate
N_detected = phi_star * A * eta

# Print results
print(f"Photon arrival rate ≈ {N_detected:.2e} photons/s")
print(f" {N_detected * 60:.2e} photons/min")
print(f" {N_detected * 3600:.2e} photons/hour")
```

Using Kepler-90’s V-band magnitude and telescope parameters, we estimate a photon detection rate of approximately 2.05e+04 photons per second with the 1.3 m telescope and 70% system efficiency. This corresponds to about photons 1.23e+06 per minute, which provides a strong enough signal for precise photometric observations during transit events. This photon arrival rate indicates that even relatively faint stars like Kepler-90 ($m_V = 13.896$) can be effectively studied with medium-sized telescopes.
