Conditional probability means how likely one event is given that another has already occurred.

```python
A_int_B6 = [x for x in A if x in B]
A_int_C6 = [x for x in A if x in C]
B_int_C6 = [x for x in B if x in C]

P_A_given_B6 = len(A_int_B6) / len(B)
P_A_given_C6 = len(A_int_C6) / len(C)
P_B_given_C6 = len(B_int_C6) / len(C)

print("P(A | B) =", P_A_given_B6)
print("P(A | C) =", P_A_given_C6)
print("P(B | C) =", P_B_given_C6)
```
