```python
Counts = [ 86, 89 ,  89 , 86,  102, 101, 83, 80, 90, 82, 85,  94, 78, 105, 83, 80, 85, 96, 90, 86, 93, 89, 82, 77, 94, 85  ]

Number = sum(Counts) # 2290

Mean = sum(Counts)/len(Counts)

print("The mean is: ", Mean)

sorted_counts = sorted(Counts)
n = len(sorted_counts)

if n % 2 == 1:  # odd length
    median_val = sorted_counts[n // 2]
else:           # even length
    median_val = (sorted_counts[n // 2 - 1] + sorted_counts[n // 2]) / 2
print("The median is: ", median_val)
# --- Mode ---
freq = {}
for c in Counts:
    freq[c] = freq.get(c, 0) + 1

max_count = max(freq.values())
mode_vals = [k for k, v in freq.items() if v == max_count]
print("The mode is: ", mode_vals)
```

```python
import math
Vari = sum((x - Mean) ** 2 for x in Counts) / (len(Counts) - 1)
std_dev = math.sqrt(Vari)
print(f" σ = {std_dev:.2f}, variance = {Vari:.2f}")
print("Rule of thumb: About 2/3 of values should fall within [mean - σ, mean + σ].")
print("(",Mean - std_dev ,",", Mean+ std_dev,")")
```

```python
lower = 80.77
upper = 95.39

# Filter counts within the interval
counts_within = [c for c in Counts if lower <= c <= upper]
print(sum(counts_within)/Number)
## with in 2/3 of values
```
