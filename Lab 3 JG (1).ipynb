### Problem 3



To determine the magnitude of the stars we are studying, we first need to _calibrate_ the sensitivity of our telescope and detector.  We want to know the value of the constant $C$ in the expression

$$
m = -2.5 \log_{10} (R) + C,
$$

where $R$ is the rate photons arrive from a standard star in counts per second, and $m$ is the magnitude of a standard.

Since we have $n$ observations of standards, we can derive $n$ estimates for the value of $C$ by using the above formula.  Breaking this down into smaller steps, we can use list comprehension and zip() to:

 1. Generate a list of values of $-2.5 \log_{10} (R)$ from our `Counts` list.  Use the `log10()` function imported above.
 2. Generate a list of the differences between the magnitudes in `Standards` and the list from step 1.
 3. Compute the average of the values in the list from step 2.

It is a good idea to print the list generated in step 2 to verify that they are all more or less the same.

```python
neg2p5logR = [-2.5 * log10(R) for R in Counts]
C_values = [m - val for m, val in zip(Standards, neg2p5logR)]
C_avg = sum(C_values) / len(C_values)
print("List of C values:", C_values)
print("Average C:", C_avg)
```

```python
C_values_rounded = [round(c, 3) for c in C_values]
C_avg_rounded = round(C_avg, 3)
print("C values (rounded):", C_values_rounded)
print("Average C (rounded):", C_avg_rounded)
```

### Problem 5

Now that we have calibrated our instrument by finding the average value of $C$, compute the magnitudes of the stars whose photon rate is tabulated in the `Starcounts` list.

 1. What is the average magnitude?
 2. What is the magnitude of the brightest star?  Of the faintest star?
 3. What is the flux ratio between them, derived from the calibrated magnitudes?
 4. What is the flux ratio, derived from the count rate?

```python
mags = [-2.5 * log10(R) + C_avg for R in Starcounts]
avg_mag = sum(mags) / len(mags)
print("Magnitudes of stars:", [round(m, 3) for m in mags])
```

```python
brightest_mag = min(mags)   
faintest_mag = max(mags)    
print("Brightest star:", round(brightest_mag ,3))
print("Faintest star:", round(faintest_mag,3))
```

```python
flux_ratio_mags = 10 ** (0.4 * (faintest_mag - brightest_mag))
print("Flux ratio (from magnitudes):", round(flux_ratio_mags, 3))
```

```python
idx_bright = mags.index(brightest_mag)
idx_faint = mags.index(faintest_mag)

R_bright = Starcounts[idx_bright]
R_faint = Starcounts[idx_faint]
flux_ratio_counts = R_bright / R_faint
print("Flux ratio (from counts):", round(flux_ratio_counts, 3))
```
